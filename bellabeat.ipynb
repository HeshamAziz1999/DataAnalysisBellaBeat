```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dailyActivity_merged.csv')
```

```python
df
```

```python
df.Id.nunique()
```

```python
cols = ["Id","ActivityDate","TotalSteps","VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes","Calories"]
df =df[cols]
```

```python
df
```

```python
df.rename(columns={'ActivityDate' : 'Date'} , inplace=True)
```

```python
df['TotalMinuts'] = df.VeryActiveMinutes+ df.FairlyActiveMinutes+ df.LightlyActiveMinutes+ df.SedentaryMinutes
```

```python
df
```

```python
df['TotalHoure']= round(df.TotalMinuts/60)
```

```python
df
```

```python
df.info()
```

```python
df.Date=pd.to_datetime(df.Date)
```

```python
import datetime as dt
df['DayOfWeek']= df.Date.dt.day_name()
```

```python
plt.figure(figsize= ((6,4)))
plt.hist(df.DayOfWeek, bins=7, color='lightskyblue', width=.6 )
plt.grid(True)

plt.xlabel('Day Of The Week')
plt.ylabel('Freq')
plt.title('Freq By Day Of The Week')

plt.show()
```
